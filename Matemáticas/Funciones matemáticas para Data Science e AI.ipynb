## Crear una función lineal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100

m = 10
b = 5

def f(x):
    return m*x + b

x = np.linspace(-10, 10, num = N)
y = f(x)


fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
N = 100

def g(x):
    return 2*x**7 - x**4 + 3*x**2 +4

x = np.linspace(-10, 10, num = N)
y = g(x)


fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
N = 100

def f(x):
    return np.cos(x)

x = np.linspace(-10, 10, num = N)
y = f(x)


fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
N = 100

def f(x):
    return np.exp(x)

x = np.linspace(-10, 10, num = N)
y = f(x)


fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
N = 10000

def f(x):
    return np.log2(x)

x = np.linspace(0.001, 10000, num = N)
y = f(x)


fig, ax = plt.subplots()
ax.plot(x, y)
```

## Funciones seccionadas

```python
def H(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x >= 0:
            y[idx] = 1
    return y


x = np.linspace(-10, 10, num = 100)

fig, ax = plt.subplots()
ax.plot(x, H(x))
```

### Funciones de activación

- Funciones lineales
- Función escalón o de Heaviside
    $$
    H(x)=
    \begin{cases}
    0 & \quad \text{when $x < 0$}\\ 
    1 & \quad \text{otherwise}
    \end{cases}
    $$
- Función sigmoide (aplicado en  regresion logistica)
    $$
    P(x) =\dfrac{1}{1+e^{-x}}
    $$
- Función tangente hiperbólica (función de escalamiento)
- Función ReLU
    $$
    R(x) = max(0, x)
    $$
