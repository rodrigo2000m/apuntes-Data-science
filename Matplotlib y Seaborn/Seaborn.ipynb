## Parámetros más utilizados

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
type(tips)
```

```python
tips.head()
```

```python
sns.displot(data=tips, x="total_bill")
```

```python
sns.displot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", hue="sex", kind="kde", legend=True,
palette="dark", alpha=0.24)
plt.show()
```

## Distribuciones

```python
sns.histplot(data=tips, x="tip", bins=15, cumulative=False, hue="sex", stat="percent",
multiple="stack")
plt.show()
```

```python
sns.kdeplot(data=tips, x="tip", hue="sex", fill=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x="tip", hue="sex")
plt.show()
```

```python
sns.displot(data=tips, x="tip", hue="sex", kind="kde")
plt.show()
```

## Categóricos

```python
sns.countplot(data=tips, x="day", hue="sex", )
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="sex")
plt.show()
```

## Relation

```python
markers= {"Lunch":"D", "Dinner":"s"}
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size="size", markers=markers)
plt.legend(loc="center", bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip", hue="day")
plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day")
plt.show()
```

## Jointplot y Pairplot

```python
sns.jointplot(data=tips, x="total_bill", y="tip", hue="sex", kind="hist", marginal_ticks=True,
marginal_kws=dict(bins=25, fill=False, multiple="dodge"))
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips, corner=True, hue="sex")
plt.show()
```
