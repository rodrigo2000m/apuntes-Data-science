# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Pyplot b√°sico

```python
x = np.linspace(0,5, 11)
y = x**2
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, "ro-") #r de red, x tipo de punto y - para indicar linea
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

## Subplot

```python
#para realizar dos gr√°ficos al mismo tiempo
plt.subplot(1, 2, 1) #los primeros dos numeros son las filas y las columnas, y el √∫ltimo n√∫mero es el gr√°fico al cual me refiero
plt.plot(x, y)
plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

```python
#para realizar dos gr√°ficos al mismo tiempo
plt.subplot(1, 2, 1) 
plt.plot(x, y, "xo-")
plt.plot(y, x, "bo-")
plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

## M√©todo orientado a objeto

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, "b")
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes.plot(x,y, "b")
axes2.plot(x,y, "r")
axes2.set_facecolor("grey")
fig.show()
```

## Subplots
Igual que subplot pero trabaja orientado a objetos

```python
fig, axes =plt.subplots(nrows=1, ncols=2, figsize= (8,5))
axes[0].plot(x, y, label="$x^2$")
axes[0].set_title("Relaci√≥n x-y")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend()

axes[1].plot(y, x, "r")
axes[1].set_title("Relaci√≥n y-x")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")
```

## Barplot

```python
country=["India", "Japon", "Mexico", "Colombia", "Alemania"]
population=[1000, 800, 900, 1000, 300]
```

```python
plt.style.use("Solarize_Light2")
plt.bar(country, population, width=0.8, color=["red", "blue"])
plt.xticks(np.arange(5), ("India", "Japon", "Mexico", "Colombia", "Alemania"), rotation=45)
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
