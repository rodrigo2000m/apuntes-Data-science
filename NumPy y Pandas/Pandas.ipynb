## Groupby
Para agrupar datos

```python
df_books.groupby("Author").count()
#df_books.groupby("Author").mean()   max()   min()   sum()
```

```python
df_books.groupby("Author").count().loc["William P. Young"]
```

```python
# si quiero que el autor no sea el indice
df_books.groupby("Author").count().reset_index()
```

```python
# me agrega los datos, me devuelve el min y el max de cada columna
df_books.groupby("Author").agg(["min", "max"])
```

```python
df_books.groupby("Author").agg({
    "Reviews":["min", "max"],
    "User Rating": "sum"
})
```

```python
df_books.groupby(["Author","Year"]).count()
```

### Concat

![concat](https://miro.medium.com/max/1400/1*NLnoAF5uOSBC2Y7IuzfM_Q.png)

```python
dict1 = {
    "A":["A0","A1","A2","A3"],
    "B":["B0","B1","B2","B3"],
    "C":["C0","C1","C2","C3"],
    "D":["D0","D1","D2","D3"]
    }
dict2 = {
    "A":["A4","A5","A6","A7"],
    "B":["B4","B5","B6","B7"],
    "C":["C4","C5","C6","C7"],
    "D":["D4","D5","D6","D7"]
    }
df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
izq = pd.DataFrame({
    "key": ["k0", "k1", "k2", "k3"],
    "A":["A0","A1","A2","A3"],
    "B":["B0","B1","B2","B3"]
    })
der = pd.DataFrame({
    "key2": ["k0", "k1", "k2", "k3"],
    "C":["C0","C1","C2","C3"],
    "D":["D0","D1","D2","D3"]
    })
```

```python
#izq.merge(der, on="key")
izq.merge(der, left_on="key", right_on="key2")
```

```python
izq.merge(der, left_on="key", right_on="key2", how="right")
```

```python
izq.join(der, how="left")
```
